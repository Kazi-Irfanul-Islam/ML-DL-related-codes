Seaborn Tutorail With Line-by-Line Example

```python
# import the librarires we need
import seaborn as sns                 # saborn for high-level data visualization
import matplotlib.pyplot as plt       # Matplotlib for lower_lever plotting(Seaborn s built on top of it)
import numpy as np                    # Numpy for numberical computations (not heavily used here, but often useful)
import pandas as pd                   #pandas for working with dataframes

# Note : Seaborn comes with some built-in datasets that can be loaded directly
```

Loading the Tips Dataset

```python
# Load the "tips" dataset - contains restaurant bills and tips
tips = sns.load_dataset("tips")

# show the first 5 rows of the dataset
tips.head()
```

Setting a Theme


```python
# set a nice theme for all plots(grid,colors, fonts etc
sns.set_theme()
```

Relational Plot

```python
# create a relational plot : total bill : tip
# col = 'time'    => seperate plots for lunch and dinner
# hue = 'smoker'  => different colors for smoker vs non-smoker
# style = 'smoker'=> different marker styles for smoker vs non-smoker
# size = 'size'   => point size varies based on party size

sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

Loading the Iris Dataset

```python
# Load the classic iris dataset(flower measurement)
iris = sns.load_dataset('iris')

# show first 5 rows
iris.head()
```

Scatter plots

```python
# scatter plot of sepal_length vs petal_length
# hue = 'spacies'         => color points by species type
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species')

# Scatter plot of sepal_length vs petal_width
sns.scatterplot(x='sepal_length', y='petal_width', data=iris, hue='species')
```

Loading the titanic dataset

```python
# Load the titanic dataset (passenger survival info)
titanic = sns.load_dataset('titanic')

#show first 5 rows
titanic.head()
```

```python
# show shape of dataset : 891 rows * 15 columns
titanic.shape
```

Count Plots

```python
# count of passengers per class(First, Second, Third)
sns.countplot(x='class', data=titanic)

# count of passengers who survived(0 = NO, 1 = YES)
sns.countplot(x='survived', data=titanic)
```

Bar Chart

```python
# Bar plot : compare survival rate across gender and class
# x-axis : male / femlae
# y-axis : survival proportion
# hue = 'class' => seperate bar for First/Second/Third Class
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

Boston Housing Dataset

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load California housing dataset
california = fetch_california_housing()

# Convert to DataFrame
house = pd.DataFrame(california.data, columns=california.feature_names)

# Add target column
house['PRICE'] = california.target

# Show first 5 rows
print(house.head())
```

diistribution Plot

```python
# Plot distribution of house prices
sns.distplot(house['PRICE'])
```

Corellation and HeatMap

```python
# Compute correlation matrix for all numberic for all numeric columns
correlation = house.corr()

# plot a heatmap of the correlation matrix
plt.figure(figsize=(10, 20))              # set figure size
sns.heatmap(correlation,                  # pass correlation matrix
            cbar=True,                    # show color bar
            square=True,                  # make cells square-shaped
            fmt='.1f',                    # format numbers to 1 decimal place
            annot=True,                   # show numbers inside cells
            annot_kws={'size':8},         # font size of annotations
            cmap='Blues',                 # color scheme(blue shades)
            )
```
