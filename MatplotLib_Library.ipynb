MatPlotLib Library

```python
# importing matplotlib library (pyplot is a module of matplotlib)
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots (used for numerical operations and array)
import numpy as np
```

```python
# creating 100 values evenly spaced between 0 to 10
x = np.linspace(0, 10, 100);

# sine values of x
y = np.sin(x);

# cosine values of x
z = np.cos(x)
```

```python
#printing the x values (from 0 to 10, divided into 100 parts)
print(x)
```

```python
# printing sine values of x
print(y)
```

```python
# printing cosine values of x
print(z)
```

Plotting basic sine and cosine wave


```python
# plot a sine wave
plt.plot(x, y)  # draw line plot of x vs y
plt.show()      # display the plot
```

```python
# plot a cosine wave
plt.plot(x, z) # draw line plot of x vs z
plt.show()     # display the plot
```

```python
# sine wave with title and axis labels
plt.plot(x, y)          # plot sine wave
plt.xlabel('angle')     # label for x-axis
plt.ylabel('sine value')# label for y-axis
plt.title('sine wave')  # title for the plot
plt.show()              # show the plot
```

Parabola Example

```python
# parabola example
x = np.linspace(-10, 10, 20)        # 20 values from -10 to 10
y = x**2;                           # sqaure of x( parabola )
plt.plot(x, y)                      # line plot
plt.show()
```

CHanging Plot Markers

```python
# parabola with green dot markers
plt.plot(x, y, 'r+')        # 'r+' means red color with plus markers
plt.show()
```

```python
# parabola with green dot markers
plt.plot(x, y, 'g.')    # 'g.' means green color with dot markers
plt.show()
```

```python
# parabola with red 'x' markers
plt.plot(x, y, 'rx')    # 'rx' means red color with 'x' markers
plt.show()
```

Multiple Line Plots on Same Graph

```python
x = np.linspace(-5, 5, 50);
plt.plot(x, np.sin(x), 'g-')   # green solid line for sine
plt.plot(x, np.cos(x), 'r--')  # red dashed line for cosine
plt.show()
```

Bar Plot

```python
# Creating a bar plot
fig = plt.figure()                                      # create a new figure
ax = fig.add_axes([0,0,1,1])                            # add axes inside the figure
language = ['English', 'French', 'Spanish', 'German']   # categories
people = [100, 30, 150, 70]                         # values for each category
ax.bar(language, people)                                # create bar plot
plt.xlabel('LANGUAGES')                                 # label x-axis
plt.ylabel('NUMBER OF PEOPLE')                          # label y-axis
plt.show()                                              # show the plot
```

Pie Chanrt

```python
# Creating a pie chart
fig1 = plt.figure();
ax = fig1.add_axes([0,0,1,1])
language = ['English', 'French', 'Spanish', 'German']
people = [100, 50, 150, 70];
ax.pie(people, labels = language, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
# scatter plot of sine and cosine values
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
#
```

3D scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Greens')
plt.show()
```
